```python
buffer = tab.loc[snap]['buffer_comovD(Mpc)']
comovMax = tab.loc[snap]['comovD_Max(Mpc)']
comovMin = tab.loc[snap]['comovD_Min(Mpc)']
ox = tab.loc[snap]['origin_x']
oy = tab.loc[snap]['origin_y']
oz = tab.loc[snap]['origin_z']
Lbox = tab.loc[snap]['BoxSize']
```

```python
@dask.delayed
def wrapper(i):
    filename = 'SubSnap_'+snap+'.'+str(i)+'.hdf5'
    input_path = os.path.join(input_dir, filename)
    with h5py.File(input_path, 'r') as f_in:
        subhalos = f_in["Subhalos"][:]
        nestedsubhalos = f_in["NestedSubhalos"][:]
        positions = subhalos["ComovingMostBoundPosition"]
        # Use a helper function to compute distance and mask
        def compute_mask(positions):
            r_comov = (positions[:,0]-ox)**2 + (positions[:,1]-oy)**2 + (positions[:,2]-oz)**2
            r_comov2 = (positions[:,0]-ox-Lbox)**2 + (positions[:,1]-oy)**2 + (positions[:,2]-oz)**2
            r_comov3 = (positions[:,0]-ox)**2 + (positions[:,1]-oy-Lbox)**2 + (positions[:,2]-oz)**2
            r_comov4 = (positions[:,0]-ox)**2 + (positions[:,1]-oy)**2 + (positions[:,2]-oz-Lbox)**2

            mask = ((r_comov < (comovMax+buffer)**2) & (r_comov > (comovMin-buffer)**2)) | \
                   ((r_comov2 < (comovMax+buffer)**2) & (r_comov2 > (comovMin-buffer)**2)) | \
                   ((r_comov3 < (comovMax+buffer)**2) & (r_comov3 > (comovMin-buffer)**2)) | \
                   ((r_comov4 < (comovMax+buffer)**2) & (r_comov4 > (comovMin-buffer)**2))
            return mask
        # Calculate the mask
        mask = compute_mask(positions)
        if np.sum(mask)==0:
            return meta
            
        # Filter both arrays using the same mask
        filtered_subhalos = subhalos[mask]
        nestedsubhalos = nestedsubhalos[mask]
        df = pd.DataFrame(filtered_subhalos[['TrackId','Nbound','Rank','Depth','Mbound','BoundM200Crit','RHalfComoving','REncloseComoving','BoundR200CritComoving','SinkTrackId','DescendantTrackId','NestedParentTrackId']],index=np.arange(len(filtered_subhalos)))
        df['InertialTensor0'] = filtered_subhalos['InertialTensor'][:,0]
        df['InertialTensor1'] = filtered_subhalos['InertialTensor'][:,1]
        df['InertialTensor2'] = filtered_subhalos['InertialTensor'][:,2]
        df['InertialTensor3'] = filtered_subhalos['InertialTensor'][:,3]
        df['InertialTensor4'] = filtered_subhalos['InertialTensor'][:,4]
        df['InertialTensor5'] = filtered_subhalos['InertialTensor'][:,5]
        df['InertialTensorWeighted0'] = filtered_subhalos['InertialTensorWeighted'][:,0]
        df['InertialTensorWeighted1'] = filtered_subhalos['InertialTensorWeighted'][:,1]
        df['InertialTensorWeighted2'] = filtered_subhalos['InertialTensorWeighted'][:,2]
        df['InertialTensorWeighted3'] = filtered_subhalos['InertialTensorWeighted'][:,3]
        df['InertialTensorWeighted4'] = filtered_subhalos['InertialTensorWeighted'][:,4]
        df['InertialTensorWeighted5'] = filtered_subhalos['InertialTensorWeighted'][:,5]
        df['ComovingAveragePosition0'] = filtered_subhalos['ComovingAveragePosition'][:,0]
        df['ComovingAveragePosition1'] = filtered_subhalos['ComovingAveragePosition'][:,1]
        df['ComovingAveragePosition2'] = filtered_subhalos['ComovingAveragePosition'][:,2]
        df['PhysicalAverageVelocity0'] = filtered_subhalos['PhysicalAverageVelocity'][:,0]
        df['PhysicalAverageVelocity1'] = filtered_subhalos['PhysicalAverageVelocity'][:,1]
        df['PhysicalAverageVelocity2'] = filtered_subhalos['PhysicalAverageVelocity'][:,2]
        df['ComovingMostBoundPosition0'] = filtered_subhalos['ComovingMostBoundPosition'][:,0]
        df['ComovingMostBoundPosition1'] = filtered_subhalos['ComovingMostBoundPosition'][:,1]
        df['ComovingMostBoundPosition2'] = filtered_subhalos['ComovingMostBoundPosition'][:,2]
        df['PhysicalMostBoundVelocity0'] = filtered_subhalos['PhysicalMostBoundVelocity'][:,0]
        df['PhysicalMostBoundVelocity1'] = filtered_subhalos['PhysicalMostBoundVelocity'][:,1]
        df['PhysicalMostBoundVelocity2'] = filtered_subhalos['PhysicalMostBoundVelocity'][:,2]        
        df['NestedSubhalos'] = [list(x) for x in nestedsubhalos]
        return df
```

```python
# Specifying the output columns upfront.
columns_to_keep = ['TrackId','Nbound','Rank','Depth','Mbound','BoundM200Crit','InertialTensor','InertialTensorWeighted','ComovingAveragePosition','PhysicalAverageVelocity','ComovingMostBoundPosition','PhysicalMostBoundVelocity','SinkTrackId','DescendantTrackId','NestedParentTrackId','NestedSubhalos']
columns_to_keep = ['TrackId','Nbound','Rank','Depth','Mbound','BoundM200Crit','RHalfComoving','REncloseComoving','BoundR200CritComoving','SinkTrackId','DescendantTrackId','NestedParentTrackId','InertialTensor0','InertialTensor1','InertialTensor2','InertialTensor3','InertialTensor4','InertialTensor5','InertialTensorWeighted0','InertialTensorWeighted1','InertialTensorWeighted2','InertialTensorWeighted3','InertialTensorWeighted4','InertialTensorWeighted5','ComovingAveragePosition0','ComovingAveragePosition1','ComovingAveragePosition2','PhysicalAverageVelocity0','PhysicalAverageVelocity1','PhysicalAverageVelocity2','ComovingMostBoundPosition0','ComovingMostBoundPosition1','ComovingMostBoundPosition2','PhysicalMostBoundVelocity0','PhysicalMostBoundVelocity1','PhysicalMostBoundVelocity2','NestedSubhalos']
meta = pd.DataFrame(columns=columns_to_keep, \
                    dtype=np.float32)
meta['TrackId'] = meta.TrackId.astype(np.int64)
meta['Nbound'] = meta.Nbound.astype(np.int64)
meta['Rank'] = meta.Rank.astype(np.int64)
meta['Depth'] = meta.Depth.astype(np.int64)
meta['NestedSubhalos'] = meta.NestedSubhalos.astype(object)
meta['SinkTrackId'] = meta.SinkTrackId.astype(np.int64)
meta['DescendantTrackId'] = meta.DescendantTrackId.astype(np.int64)
meta['NestedParentTrackId'] = meta.NestedParentTrackId.astype(np.int64)
```

```python
import pyarrow as pa

schema = pa.schema([
    ('TrackId', pa.int64()),
    ('Nbound', pa.int64()),
    ('Rank', pa.int64()),
    ('Depth', pa.int64()),
    ('Mbound',pa.float32()),
    ('BoundM200Crit',pa.float32()),
    ('RHalfComoving',pa.float32()),
    ('REncloseComoving',pa.float32()),
    ('BoundR200CritComoving',pa.float32()),
    ('SinkTrackId', pa.int64()),
    ('DescendantTrackId', pa.int64()),
    ('NestedParentTrackId', pa.int64()),
    ('InertialTensor0', pa.float32()),
    ('InertialTensor1', pa.float32()),
    ('InertialTensor2', pa.float32()),
    ('InertialTensor3', pa.float32()),
    ('InertialTensor4', pa.float32()),
    ('InertialTensor5', pa.float32()),
    ('InertialTensorWeighted0', pa.float32()),
    ('InertialTensorWeighted1', pa.float32()),
    ('InertialTensorWeighted2', pa.float32()),
    ('InertialTensorWeighted3', pa.float32()),
    ('InertialTensorWeighted4', pa.float32()),
    ('InertialTensorWeighted5', pa.float32()),
    ('ComovingAveragePosition0',pa.float32()),
    ('ComovingAveragePosition1',pa.float32()),
    ('ComovingAveragePosition2',pa.float32()),
    ('PhysicalAverageVelocity0',pa.float32()),
    ('PhysicalAverageVelocity1',pa.float32()),
    ('PhysicalAverageVelocity2',pa.float32()),
    ('ComovingMostBoundPosition0',pa.float32()),
    ('ComovingMostBoundPosition1',pa.float32()),
    ('ComovingMostBoundPosition2',pa.float32()),
    ('PhysicalMostBoundVelocity0',pa.float32()),
    ('PhysicalMostBoundVelocity1',pa.float32()),
    ('PhysicalMostBoundVelocity2',pa.float32()),
    ('NestedSubhalos',pa.list_(pa.int64()))
])
```

```python
t1 = time.time()
res.to_parquet(d_out / f'SubSnap_{snap}.pq',engine="pyarrow", schema=schema)
print('time', time.time() - t1)
```
